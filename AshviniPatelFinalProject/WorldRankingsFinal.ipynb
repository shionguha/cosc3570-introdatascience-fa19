```python
import pandas as pd
import datetime
from os import listdir
from os.path import isfile, join
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

import warnings 
warnings.filterwarnings("ignore")


df = pd.read_csv("Data/World University Ranks.csv");
df.head(10)


#filter on the year 2015 to condense the data
df2015 = df[df['year']==2015]
df2015 = df2015[df2015['world_rank']<101]
df2015.head()
```

```python
df = sns.regplot(x="world_rank", y="publications", data=df2015, color = "skyblue")
x=df.get_lines()[0].get_xdata()
y=df.get_lines()[0].get_ydata()
linregress(x,y)
```
