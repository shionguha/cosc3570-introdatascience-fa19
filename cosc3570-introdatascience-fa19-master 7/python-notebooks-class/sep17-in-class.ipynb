```python
df = pd.read_csv("../data/tb-raw.csv")
df
```

```python
df = pd.melt(df, id_vars=["country","year"], value_name="cases", var_name="gender_and_age")

# Extract Sex, Age lower bound and Age upper bound group
tmp_df = df["gender_and_age"].str.extract("(\D)(\d+)(\d{2})")    

# Name columns
tmp_df.columns = ["gender", "age_lower", "age_upper"]

# Create `age`column based on `age_lower` and `age_upper`
tmp_df["age"] = tmp_df["age_lower"] + "-" + tmp_df["age_upper"]

# Merge 
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns and rows
df = df.drop(["gender_and_age","age_lower","age_upper"], axis=1)
df = df.dropna()
#df = df.sort(ascending=True,columns=["country", "year", "gender", "age"])
df.head(10)
```
